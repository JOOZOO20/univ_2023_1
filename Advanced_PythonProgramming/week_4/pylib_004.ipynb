```python
import functools

def xy_compare(n1, n2):
    if n1[1] > n2[1]:         # y 좌표가 크면
        return 1
    elif n1[1] == n2[1]:      # y 좌표가 같으면
        if n1[0] > n2[0]:     # x 좌표가 크면
            return 1
        elif n1[0] == n2[0]:  # x 좌표가 같으면
            return 0
        else:                 # x 좌표가 작으면
            return -1
    else:                     # y 좌표가 작으면
        return -1

src = [(0, 4), (1, 2), (1, -1), (2, 2), (3, 3)]
result = sorted(src, key=functools.cmp_to_key(xy_compare))
print(result)
```

```python
import time

def elapsed(original_func):
  def wrapper(*args,**kwargs):
    s=time.time()
    rst=original_func(*args,**kwargs)
    e=time.time()
    print('elapsed itme: %f 초' %(e-s))
    return result
  return wrapper

@elapsed
def add(a,b):
  return a+b

@elapsed
def factorial(n):
  if n==1:
    return 1
  else:
    return n*factorial(n-1)


print(add(3,4))
print(factorial(10))
```

```python
import time
import functools

def elapsed(original_func):
  @functools.wraps(original_func)
  def wrapper(*args,**kwargs):
    s=time.time()
    rst=original_func(*args,**kwargs)
    e=time.time()
    print('elapsed itme: %f 초' %(e-s))
    return result
  return wrapper

@elapsed
def add(a,b):
  return a+b
'''두 수 a,b의 덧셈 결과를 반환하는 함수'''

print(add(3,4))
```
